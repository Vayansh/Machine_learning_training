# **By Scratch**

```python
class PCA:
  def __init__(self,n_components):
    self.n_components = n_components
    self.components = None
    self.mean = None

  def fit(self,X):
    # find the mean of x and subtract it from X
    self.mean = np.mean(X,axis = 0)
    X = X - self.mean

    # Find the covariance of X
    conv = np.cov(X,rowvar = False)

    # find eigenvector and eigenvalues of x
    eigenvalues,eigenvector = np.linalg.eig(conv)

    # sort eigenvector
    idx = np.argsort(eigenvalues)[::-1]
    eigenvector = eigenvector[:,idx]
    eigenvalues = eigenvalues[idx]
    self.components = eigenvector[:,:self.n_components]

  def transform(self,X):
    X = X - self.mean
    return np.dot(X,self.components)
```
